```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
complaints_dec14 = pd.read_csv('complaints_dec_2014.csv')
```

```python
complaints_dec14['Product'].value_counts().plot("bar")
plt.title("Number of complaints for each Product type")
plt.ylabel("Number of Complaints")
plt.show()
```

Here we see that Mortgages and Debt Collection receive the most complaints.

```python
complaints_dec14['Company'].value_counts()[:10].plot("bar")
plt.title("Topy 10 Companies for Complaints Received")
plt.ylabel("Number of Complaints")
plt.show()
```

Here we see that Bank of America recieves the most complaints, edging out all three major credit-reporting companies for the top of the top 4.

```python
complaints_dec14['Company response'].value_counts().plot("bar")
plt.title("Number of Complaints for Each Possible Company Response")
plt.ylabel("Number of Complaints")
plt.show()
```

Here we see that a large majority of complaints are closed with an explanation.

```python
complaints_dec14.index = pd.to_datetime(complaints_dec14['Date received'], '%m/%d/%Y')
```

```python
complaints_dec14['weekday'] = complaints_dec14.index.weekday
#complaints_dec14['weekday'] = complaints_dec14.index.to_series().map(lambda d: d.weekday())
```

```python
means = [complaints_dec14[complaints_dec14.weekday == x].count().Product/len(complaints_dec14[complaints_dec14.weekday == x].index.unique()) for x in range(7)]
print(means)
plt.bar(range(7), means)
plt.title('Mean Number of Complaints submitted per week day')
#plt.xticks(['M', 'Tu', 'W', 'Th', 'F', 'Sa', 'Su'])
plt.show()
```

Here we see that the beginning of the week is more likely to result in a complaint Tuesday having a slight advantage.  As could be anticipated, the weekend sees a significant decrease.
